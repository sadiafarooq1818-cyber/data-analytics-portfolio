# Retail Sales Analytics Notebook

This notebook demonstrates a medium-size data analyst workflow: load data, clean data, calculate KPIs, group results and prepare outputs for Excel or Power BI dashboards.

```python
import pandas as pd
df = pd.read_csv('../project-01-retail-sales-excel-dashboard/data/retail_sales_sample.csv')
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.to_period('M').astype(str)
df['profit_margin'] = df['profit'] / df['sales']
df.isna().sum()
```

```python
kpis = {
    'total_sales': df['sales'].sum(),
    'total_profit': df['profit'].sum(),
    'profit_margin': df['profit'].sum() / df['sales'].sum(),
    'total_orders': df['order_id'].nunique(),
    'average_order_value': df['sales'].sum() / df['order_id'].nunique()
}
kpis
```

```python
region_summary = df.groupby('region').agg(total_sales=('sales','sum'), total_profit=('profit','sum'), orders=('order_id','nunique')).sort_values('total_sales', ascending=False)
region_summary
```

```python
category_summary = df.groupby('product_category').agg(total_sales=('sales','sum'), total_profit=('profit','sum'), avg_discount=('discount','mean')).sort_values('total_sales', ascending=False)
category_summary
```
